<h2>Data analysis and plotting</h2>

<h4  style="color:#4697E8";>IV sweeped in power and frequency</h4>

```python
pwr_arr = np.unique(df_clean['mw_source_power'])
```

```python
# plt.figure(figsize=(10,10))

PWR = pwr_arr[-5]
fmax = 5e9

f_df = df_clean[(df_clean['mw_source_power'] == PWR) 
                & (df_clean['mw_source_frequency'] < fmax) ]
                # & (df_clean['adwin_input_2_voltage_up'] > 0.02)]

freq = f_df['mw_source_frequency'].to_numpy()
V = f_df['adwin_input_2_voltage_up'].to_numpy()
current= f_df['adwin_up_ramp'].to_numpy()

m_size = (np.unique(freq).size, -1)
freq = np.reshape(freq, m_size)
V = np.reshape(V, m_size)
current = np.reshape(current, m_size)
```

```python
plt.figure(figsize=(12,10))
# grd = np.gradient(current, V[0,0]-V[0,1])[1]


plt.contourf(V, freq,current, levels=150)
plt.xlabel('V')
plt.ylabel('frequency')
plt.colorbar()
```

```python
e=1.6e-19
@interact 
# def test(c1 = widgets.SelectionSlider(options=np.unique(f_df['mw_source_frequency'])*1e-9)):
def test(c1 = np.round(np.unique(f_df['mw_source_frequency'])*1e-9, decimals=4) ):
    qdf = f_df[np.isclose(f_df['mw_source_frequency']*1e-9, c1, atol = 1e-4)]
    
    plt.figure(figsize=(12,10))
    plt.plot(qdf['adwin_input_2_voltage_up'], qdf['adwin_up_ramp'])
    plt.hlines([2*e*c1*1e9*k for k in range(-1,4)], -0.15, 0.15, color='green')
```
