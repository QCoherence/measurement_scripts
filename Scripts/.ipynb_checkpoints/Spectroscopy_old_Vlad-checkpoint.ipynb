```python

#####_Extracting data###########################################################
if First_time:
    ###########################################################
    # tt_data      -- colorplot_vs_clux twotone
    # tt_fit_data  -- 1D freq_vs_flux twotone
    ###########################################################

        ## Extraction from 2D colorplot twotone
    tt_freq     = np.unique(tt_data[:,0]) #freq of second tone
    n           = len(tt_freq)
    tt_flux     = tt_data[::n, 1]
    tt_Ampl     = tt_data[:,2]
    tt_S21_dB   = np.transpose(np.reshape(tt_Ampl, (len(tt_flux), n)))

        ## Extraction from 2D onetone
    ot_freq     = np.unique(ot_data[:,0]) #freq of second tone
    n           = len(ot_freq)
    ot_flux     = ot_data[::n, 1]
    ot_current  = ot_data[::n,3]
    ot_Ampl     = ot_data[:,2]
    ot_S21_dB   = np.transpose(np.reshape(ot_Ampl, (len(ot_flux), n)))

        ## Extraction from 1D fit twotone
    tt_fit_flux = tt_fit_data[:,0]
    tt_fit_freq = tt_fit_data[:,1]

        ## Extraction from 1D fit onetone
    ot_fit_flux = ot_fit_data[:,1]
    ot_fit_upol = ot_fit_data[:,2]
    ot_fit_lpol = ot_fit_data[:,3]

        ## Extraction fit points of Remy's theory
    th_flux     = th_data[:,0]
    th_qub      = th_data[:,1]
    th_lpol     = th_data[:,2]
    th_upol     = th_data[:,3]
```

```python
#########_Erase_data_###########################################################
if First_time:
    print 'Cleaning...'

    ##========================##
    ##--Sorry for shity code--##
    ##========================##

    ###### for one tone erase frequences less than 6.5 GHz ####
    ot_freq = ot_freq[500::]
    ot_S21_dB = ot_S21_dB[500::,:]
    ################################

    ###### for two tone erase gaps ####

    dist = 154  # distance between integer number of flux in points
    zero = 226  #Number of point of tt_flux = 0

    #----------------------
    # arg = []
    # for i in np.arange(-1,5+1):
    #     arg.append(zero + i*dist -gap)
    #     arg.append(zero + i*dist +gap)
    arg1  = zero - dist -70
    arg2  = zero - dist +72
    arg3  = zero -71
    arg4  = zero +71
    arg5  = zero + dist -71
    arg6  = zero + dist +72
    arg7  = zero + 2*dist -71
    arg8  = zero + 2*dist +70
    arg9  = zero + 3*dist -71
    arg10 = zero + 3*dist +71
    arg11 = zero + 4*dist -70
    arg12 = zero + 4*dist +70
    arg13 = zero + 5*dist -70
    arg14 = zero + 5*dist +70
    #----------------------

    #==============================
    tt_flux_1 = tt_flux[arg1:arg2]
    tt_flux_2 = tt_flux[arg3:arg4]
    tt_flux_3 = tt_flux[arg5:arg6]
    tt_flux_4 = tt_flux[arg7:arg8]
    tt_flux_5 = tt_flux[arg9:arg10]
    tt_flux_6 = tt_flux[arg11:arg12]
    tt_flux_7 = tt_flux[arg13:arg14]
    tt_S21_1  = tt_S21_dB[:, arg1:arg2]
    tt_S21_2  = tt_S21_dB[:, arg3:arg4]
    tt_S21_3  = tt_S21_dB[:, arg5:arg6]
    tt_S21_4  = tt_S21_dB[:, arg7:arg8]
    tt_S21_5  = tt_S21_dB[:, arg9:arg10]
    tt_S21_6  = tt_S21_dB[:, arg11:arg12]
    tt_S21_7  = tt_S21_dB[:, arg13:arg14]
    #==============================

    #==============================
    ## extracted points of twotone
    bb = 592
    tt_fit_freq = np.delete(tt_fit_freq, bb)
    tt_fit_flux = np.delete(tt_fit_flux, bb)

    ## extracted points for Onetone
    bb = 719
    ot_fit_lpol = np.delete(ot_fit_lpol, bb)
    ot_fit_upol = np.delete(ot_fit_upol, bb)
    ot_fit_flux = np.delete(ot_fit_flux, bb)
```
