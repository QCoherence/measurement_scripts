```python
def powtov(power) :
    poww = 10**(power/10)/1000
    v = np.sqrt(50*poww*2)
    return v
def vtopow(v) :
    poww = v**2/(100)
    power = 10*np.log10(poww)+30
    return power
```

```python
def meas_with_RfSoc(LO_freq_vec, exp, station):
    meas = qc.Measurement(exp=exp, station=station)
    # meas.register_parameter(source_LO.freq_vec)
    meas.register_parameter(rfsoc.IQINT_ALL)
    #                         setpoints=[source_LO.frequency])

    # source_LO.status('on')
    with meas.run() as datasaver:
        id=datasaver.dataset.run_id
        qc.load_by_run_spec( captured_run_id=id).add_metadata('parameter_snap',
                             json.dumps(parameter_snap))
        res=rfsoc.IQINT_ALL.get()

        # I = np.average(res[1][0].reshape(len(freq_vec), average), axis=1)
        # Q = np.average(res[0][0].reshape(len(freq_vec), average), axis=1)

        # datasaver.add_result((rfsoc.IQINT_ALL,[I,Q]),(source_LO.freq_vec, LO_freq_vec))

        
    return res
```

```python
###############################################################################
#                             Parameter RfSoc
###############################################################################
ch_demod = [3]

amp_sig = powtov(0) #6dbm
amp_sig = 0.9
P_in = vtopow(amp_sig)

freq_sig = 100 #100 MHz
phase_sig = 0
phase_off = 0
dc_off = 0

acq_length = 10 #Âµs

time_shift = 270e-3
time_start_trigger = 0
time_start_dac = 1e3
time_start_dac = 0.1e3
# time_start_dac = 0
time_start_adc =  time_shift + time_start_dac
dead_time = 10
trigger_length = 1


average = int(3)
###############################################################################
#                             Parameter Lo Source
###############################################################################
source_power = 22
# source_power = 0

freq_start  = 7.4 + freq_sig*1e-3
freq_end    = 7.7 + freq_sig*1e-3
points_freq = 10
freq_step   = (freq_end - freq_start)/points_freq
freq_vec    = np.linspace(freq_start, freq_end, points_freq)
```

```python

##############################################################################
# Fill table 
##############################################################################

param_sin_1 = {'amp':1,
             'freq':freq_sig,
             'dc_offset':dc_off,
             'phase_offset':0*np.pi/180}

param_sin_2 = {'amp':1,
             'freq':freq_sig,
             'dc_offset':dc_off,
             'phase_offset':90*np.pi/180}

param_sin_3 = {'amp':1,
             'freq':freq_sig,
             'dc_offset':dc_off,
             'phase_offset':180*np.pi/180}

param_sin_4 = {'amp':1,
             'freq':freq_sig,
             'dc_offset':dc_off,
             'phase_offset':270*np.pi/180}

param_sin_I= {'amp':amp_sig,
             'freq':freq_sig,
             'dc_offset':dc_off,
             'phase_offset':0*np.pi/180}





pulse_sin = dict(label='signal', 
                      module='DAC', 
                      channel=3, 
                      mode='sin', 
                      start=time_start_dac, 
                      length=acq_length,
                      dead_time = dead_time,
                      repetitions=average,
                      param=param_sin_I, 
                      LUT=True,
                      starting_pointer=0,
                      parent=None)

pulse_wait = dict(label='wait_after_trigger',
                      module='DAC', 
                      channel=3, 
                      mode='DC',
                      start=time_start_trigger + trigger_length,
                      length = time_start_dac - (time_start_trigger + trigger_length),
                      repetitions=1, 
                      param={'amp':0},
                      LUT=True,
                      parent=None)



record_sin = dict(label='record_signal', 
                      module='ADC', 
                      channel=3, 
                      demodulation_channels=ch_demod,
                      LUT=True, 
                      function ='sin',
                      param=[param_sin_1, param_sin_2, param_sin_3, param_sin_4],
                      mode='IQ_table', 
                      start=time_start_adc, 
                      length=acq_length - 2*time_shift, 
                      dead_time = dead_time,
                      repetitions=average,
                      parent=None)

trigger_pulse = dict(label='trigger', 
                     module='DAC', 
                     channel=3, 
                     mode='trigger', 
                     start =time_start_trigger, 
                     length=trigger_length, 
                     repetitions=1,
                     LUT=True,
                     starting_pointer=0,
                     param=None, 
                     parent=None)



parameter_snap['RfSoc'] = {'amp_sig' : amp_sig, 'freq_sig': freq_sig, 
                          'phase_sig': phase_sig, 'offset' : [dc_off,phase_off],
                          'start_sig': time_start_adc, 'sig_length' : acq_length,
                          'record_start': time_start_dac}


pulses = pd.DataFrame()
pulses = pd.concat([pulses, pd.DataFrame.from_records([trigger_pulse])], ignore_index=True)
pulses = pd.concat([pulses, pd.DataFrame.from_records([pulse_wait])], ignore_index=True)  
pulses = pd.concat([pulses, pd.DataFrame.from_records([pulse_sin])], ignore_index=True)
pulses = pd.concat([pulses, pd.DataFrame.from_records([record_sin])], ignore_index=True)

rfsoc.pulses = pulses

#     ###############################################################################
#     #                             Setting up microwave source 
#     ###############################################################################

#     if sweep: 
#         source_LO.power(source_power)
#         source_LO.sourcemode('sweep')
#         source_LO.sweepmode('STEP')
#         source_LO.trigger_mode('EXT')
#         source_LO.spacing_freq('LIN')
#         source_LO.freq_start((freq_start-freq_step)*1e9) # Hack to take into account that the first
#                                            # frequency point of MW_source doesn't send
#                                            # a trigger to the VNA
#         source_LO.freq_stop(freq_end*1e9)
#         source_LO.freq_step(freq_step*1e9)
#         source_LO.freq_points(points_freq+1)
#         source_LO.set_gui_update('on')
#         source_LO.restartsweep()
#         time.sleep(0.5)
#         # time.sleep(5)
#         source_LO.set_gui_update('off')

#     else: 
#         source_LO.sourcemode('CW')
#         source_LO.power(source_power)
#         source_LO.frequency(int(freq))
```
