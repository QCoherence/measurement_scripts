## Sweep VNA power and acquire normalized VNA trace

```python
parameter_snap={}

# VNA
start_freq= 6.0
end_freq= 6.3
points_per_gigahz = 1000
VNA_power_start = -20
VNA_power_end = 10
VNA_power_step = 0.2
IF_bandwidth=100
Averages=1
points_VNA=1+(end_freq-start_freq)*points_per_gigahz
points_VNA=int(points_VNA)
points_VNA_power=1+(VNA_power_end-VNA_power_start)/VNA_power_step
points_VNA_power=int(points_VNA_power)
parameter_snap['vna']={'start_freq':start_freq,'end_freq':end_freq,'points_per_gigahz':points_per_gigahz,'VNA_power_start':VNA_power_start,'VNA_power_end':VNA_power_end,'VNA_power_step':VNA_power_step,'IF_bandwidth':IF_bandwidth,'Averages':Averages,'points':points_VNA,'points_VNA_power':points_VNA_power}

# Current source
I = 0
parameter_snap['flux']={'I':I}


vna_tvp.channels.S21.bandwidth(IF_bandwidth)
vna_tvp.channels.S21.power(VNA_power_start)
vna_tvp.channels.S21.avg(Averages)
vna_tvp.channels.npts(points_VNA)
vna_tvp.channels.start(start_freq*1e9)
vna_tvp.channels.stop(end_freq*1e9)



# Initialize devices and measurement
vna_tvp.rf_on()
vna_tvp.cont_meas_on()
current_source_on()
current_set(I)
MW_source_1_tvp.status('off')
# MW_source_2_tvp.status('off')
# set_rt_switch('NC')

exp_name = 'Transmission vs VNA power'
# sample_name = 'T3PA V08 02'
sample_name = 'JPA graphene'
exp=qc.load_or_create_experiment(experiment_name=exp_name,
                          sample_name=sample_name)

meas = qc.Measurement(exp=exp, station=station) 
meas.register_parameter(vna_tvp.channels.S21.power) 
meas.register_parameter(vna_tvp.channels.S21.trace_mag_phase, setpoints=(vna_tvp.channels.S21.power,)) 
meas.write_period = 2

with meas.run() as datasaver:
    for power in bar(np.linspace(VNA_power_start, VNA_power_end, points_VNA_power)):
        vna_tvp.channels.S21.power(power)
        get_v = vna_tvp.channels.S21.trace_mag_phase.get()
        datasaver.add_result((vna_tvp.channels.S21.power, power),
                             (vna_tvp.channels.S21.trace_mag_phase, get_v))
            

            
# Shutdown devices
vna_tvp.rf_off()
vna_tvp.cont_meas_off()
current_source_off()
# snotify('arpit.ranadive@gmail.com','Finished: '+exp_name+' '+sample_name)
```

```python
captured_run_id = 5
dataset = qc.load_by_run_spec(captured_run_id=captured_run_id)
dataset.add_metadata('parameter_snap',json.dumps(parameter_snap))
# json.loads(dataset.get_metadata('parameter_snap'))
```

```python
power_data=list(dict.fromkeys(dataset.get_parameter_data('VNA_S21_power')['VNA_S21_power']['VNA_S21_power']))
freq_data=dataset.get_parameter_data('magnitude')['magnitude']['VNA_S21_S21_frequency'][0:points_VNA]*1e-9
magnitude=(dataset.get_parameter_data('magnitude')['magnitude']['magnitude']).reshape(len(power_data),len(freq_data))
```

```python
plot_title=database_name+'\n'+exp_name+'  ['+sample_name+']  (id: '+str(captured_run_id)+')'
plot_title+='\n Freq: '+str(start_freq)+' to '+str(end_freq)+';'+str(points_per_gigahz)+'    VNA_power: '+str(VNA_power_start)+' to '+str(VNA_power_end)+';'+str(VNA_power_step)+'    I: '+str(I)+' mA'

plt.figure(figsize=(14, 10), dpi= 80) #facecolor='w', edgecolor='k'

X, Y = np.meshgrid(np.array(power_data),np.array(freq_data))
Z = magnitude.T
plt.pcolor(X, Y, Z)
# plt.pcolor(X, Y, Z, vmin=0, vmax=-30)
cbar=plt.colorbar()
cbar.set_label('Transmission [dB]', size=14)

plt.rc('axes', labelsize=12)    # fontsize of the x and y labels

x_label='VNA_power [dBm]'
y_label='Frequency [GHz]'
plt.xlabel(x_label, size=14)
plt.ylabel(y_label, size=14)
plt.title(plot_title, size=16)

plt.show()
```
