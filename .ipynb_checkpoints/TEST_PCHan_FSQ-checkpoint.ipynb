## Integration Bandwidth Method (IBM)

```python
FSQ_center = 3.              #GHz
FSQ_span   = 100.*1.e-6      #GHz
FSQ_avg    = 1
FSQ_rbw    = FSQ_span * 0.01 #GHz #for optimal selectivity of the channel bandwidth (avoid adjacent channel's power to be added to the measured channel)
FSQ_vbw    = 3*FSQ_rbw       #GHz #for optimal use of the RMS detector

FSQ_npt    = int(FSQ_span/FSQ_rbw) + 1

FSQ_ref    = 0
FSQ_att    = 0

FSQ_sweeptime = 1./(FSQ_vbw*1.e9) * FSQ_npt * 10


```

```python
FSQ.res_BW(FSQ_rbw*1.e9)
FSQ.video_BW(FSQ_vbw*1.e9)
FSQ.center_freq(FSQ_center*1.e9)
FSQ.averages(FSQ_avg)
# FSQ.n_points(FSQ_npt)
# FSQ.span(FSQ_span*1.e9)
FSQ.ref_level(FSQ_ref)
FSQ.input_att(FSQ_att)
FSQ.detector('rms')
FSQ.sweep_time_direct(FSQ_sweeptime)
```

```python
FSQ.write('SENSe1:POW:ACH:ACP 0') # pure channel measurements
FSQ.write('SENSe1:POW:ACH:BAND {}'.format(FSQ_span*1.e9))
FSQ.write('SENSe1:POW:ACH:MODE ABS')
FSQ.write('CALC1:MARK:FUNC:POW:SEL CPOW') 
FSQ.write('INIT:CONT OFF')
FSQ.write('CALC1:MARK:FUNC:POW:RES:PHZ OFF')

N=100

data=np.zeros(N)

for i in range(N):
    FSQ.write('INIT;*WAI')
    data[i]=FSQ.ask('CALC1:MARK:FUNC:POW:RES? CPOW')
```
