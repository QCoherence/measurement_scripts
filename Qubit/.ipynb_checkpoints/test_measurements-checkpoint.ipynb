<h2>Spectroscopy</h2>
The two modes of the arrays to be used to pulse and readout are analysed in frequency and phase.

```python
from sweep import Sweep

f1 = 3.884e9
f2 = 4.003e9
```

<h4>Mode 1</h4>

```python
spectroscopy_1 = Sweep( freq_center = f1,
                        freq_span = 20e6,
                        df = 20e3,
                        num_averages = 50,
                        amp = 0.1,
                        output_port = 1,
                        input_port = 1 
                       )
```

```python
save_filename_1 = spectroscopy_1.run(ADDRESS)
spectroscopy_1.analyze();
```

<h4>Mode 2</h4>

```python
spectroscopy_2 = Sweep( freq_center = f2,
                        freq_span = 20e6,
                        df = 20e3,
                        num_averages = 10,
                        amp = 0.2,
                        output_port = 1,
                        input_port = 1 
                      )
```

```python
save_filename_2 = spectroscopy_2.run(ADDRESS)
spectroscopy_2.analyze();
```

<h2>Ramsey measurement</h2>
Perform a Ramsey-type spectroscopy for the measurement of the mode's coherence time.

```python
from ramsey_single import RamseySingle
```

```python
delay_start = 0
delay_stop = 0.2e-6
delay_points = 101
delays = np.linspace(delay_start, delay_stop, delay_points)

ramsey = RamseySingle(  readout_freq = f1,
                        control_freq = f2,
                        readout_amp = 0.1,
                        control_amp = 0.1,
                        readout_duration = 50e-9,
                        control_duration = 50e-9,
                        sample_duration = 50e-9,
                        delay_arr = delays,
                        readout_port = 1,
                        control_port = 1,
                        sample_port = 1,
                        wait_delay = 5e-6,
                        readout_sample_delay = 5e-6,
                        num_averages = 10001
                      )
```

```python
ramsey.run(ADDRESS)
ramsey.analyze()
```
